###2.Control del flujo, Iteradores y Funciones

####2.1 Control de flujo
Una estructura de control, es un bloque de código que permite agrupar instrucciones de manera controlada. 
Las estructuras de control de flujo condicionales, se definen mediante el uso de tres palabras claves reservadas, del lenguaje: `if` (si), `elif` (sino, si) y `else` (sino).

````python
a = 3
b = 2
#
if a < b: 
    print('se cumplió la condición 1')
elif a == b:
    print('no se cumplió la condición 1, pero si la 2')
else:
    print('no se cumplió ninguna de las condición anteriores')
````

#####For
La instrucción `for` itera con una variable que toma valores de una lista dada. A menudo, esa lista se especifica mediante `range`.

````python
#se puede iterar sobre los valores de una lista
lista = [1,2,3,4,5,6,7,8]
for var in lista:
  numero = var **2
  print(numero)
````

````python
list(range(len(lista)))
````

````python
#se puede iterar sobre los indices de una lista
for i in range(len(lista)):
  print(i)
````

````python
#Se puede iterar sobre una lista
lista2 = []

for i in lista:
  numero2 = i**3
  lista2.append(numero2)
print(lista2)
````

````python
#se pueden incorporar controles de flujo
lista2 = []
print(lista2)
for i in lista:
  numero = i**2
  if numero%2 == 0:
      lista2.append(numero)
print(lista2)
````

````python
#List comprehension
l2 = [i**2 for i in lista]
print(l2)
````

````python
li = [i-1 for i in l2]
print(li)
````

````python
#mas ejemplos de list comprehension

l2 = [i**2 for i in lista if i%2 == 0]
print(l2)

l2 = [1 if i%2==0 else 0 for i in lista]
print(l2)
````

#####While

La instrucción `While` itera con una variable entregada que va mutando *dentro*. Requiere partir de una condicion verdadera

````python
#, es decir 'while (condition) ==False'
i=20
while (i<10):
    print('iteración:', i)
    i+=1
````

Veamos un ejemplo "util":



Si tenemos el polinomio:

$$
\begin{align}
P(x) &= 5+2x-3x^2+4x^3
\end{align}
$$
Se puede factorizar como:
$$
\begin{align}
P(x) &=5+x(2+x(-3+x(4)))
\end{align}
$$


````python
a = [5,2,-3,4]
x = 2

P=0
print(P)
i = len(a)
while i > 1:
  P = (P + a[i-1]) * x
  print(P)
  i -= 1
P += a[0]
print(P)
````

Con **global** las declaraciones dentro de la función al ejecutarse pueden cambiar al valor original de la variable definida a fuera de la función. En general no se recomiendo el uso de "global".

##Programación orientada a objetos (OOP)

**Comentario 5:** *quizas ver algo de programacion orientada a objetos (OOP)? aunque nosé si se salga mucho del enfoque del curso jeje* 

1. Clase: Propiedad y comportamiento de un objeto concreto.
2. Objeto: Instancia de una clase
3. Atributo: Propiedad del objeto
4. Método: Lo que un objeto puede hacer 



Más documentación<br><br>
Métodos especiales: https://docs.python.org/3/reference/datamodel.html#special-method-names<br>


````python
class Book():
  
    #Método constructor: Función para instanciar un objeto. 
    def __init__(self, title, author, pages): 
        print("A book is created") #mensaje que se muestra en patalla al crear un objeto
        #Atributos de cada objeto
        self.title = title 
        self.author = author
        self.pages = pages
    #Métodos especiales
    def __str__(self): #al utilizar la función print() sobre un objeto retorna un string con todos sus atributos de forma. sin este método al imprimir un objeto este se 
        return "Title:%s , author:%s, pages:%s " %(self.title, self.author, self.pages)
    def __len__(self):#al utilizar la función len() sobre un objeto retorna el número de páginas del libro
        return self.pages 
    #método    
    def saludar(self, saludo):
        return '{s} {autor}'.format(s=saludo, autor=self.author)
````

````python
# instanciar un objeto
book = Book("Machine Learning A Probabilistic Perspective", "K. Murphy", 1000) 
````

````python
book
````

````python
# acceder a un atributo de un objeto
book.title
````

````python
# método especial: print
print(book)
````

````python
# método especial:len
len(book)
````

````python
# método normal
book.saludar('Hola')
````

````python
# importar las librerias a utilizar
import numpy as np
import pandas as pd
````

````python
etiquetas = ['a', 'b', 'c']
datos = [1,2,3]
serie_1 = pd.Series(data = datos, index  = etiquetas)
serie_1
````

####Dataframes

````python
#Podemos crear nuestro DF
#Opcion 1
filas = ['ventas1', 'ventas2', 'ventas 3']
columnas = ['zonaA', 'zonaB', 'zonaC']
datos = [[123,421,256], [234,541,257], [120,451,258]]

dataframe = pd.DataFrame(datos, filas, columnas)
````

````python
#Opcion 2
dic_ventas = {
    'zonaA':[123,421,256], 
    'zonaB':[234,541,257], 
    'zonaC':[120,451,258]
}

dataframe_2 = pd.DataFrame(data=dic_ventas,index=filas)
dataframe_2
````
